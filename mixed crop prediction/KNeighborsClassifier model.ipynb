```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.multioutput import MultiOutputClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Load data
data = pd.read_csv("mixedcrop.csv")
data
```

```python
# Split data into features (x) and labels (y)
x = data.iloc[:, :-2]
y = data.iloc[:, [7, 8]]

# Split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=0)

# Standardize features
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
x_train
x_test
y_test
```

```python

# Encode labels
lab = LabelEncoder()
lab1 = LabelEncoder()
data["label"] = lab.fit_transform(data["label"])
data["crop"] = lab1.fit_transform(data["crop"])
data["label"] 
data["crop"]
```

```python

# Create KNeighborsClassifier model
knn = KNeighborsClassifier(n_neighbors=3,metric="manhattan")
model = MultiOutputClassifier(knn)

# Fit the model
model.fit(x_train, y_train)

# Predict on the test set
y_pred = model.predict(x_test)
y_pred
```

```python

# Calculate accuracy for each output
accuracy_label = accuracy_score(y_test["label"], y_pred[:, 0])
accuracy_crop = accuracy_score(y_test["crop"], y_pred[:, 1])

print("Accuracy for crop1:", accuracy_label)
print("Accuracy for crop2:", accuracy_crop)
```

```python

# Take user input for prediction
a = []
for i in range(7):
    a.append(float(input()))
a = np.array(a)

# Make predictions on user input
prediction = model.predict(a.reshape(1, -1))

c1 = prediction[0, 0]
c2 = prediction[0, 1]

print("Predicted crop1:", c1)
print("Predicted crop2:", c2)
```
